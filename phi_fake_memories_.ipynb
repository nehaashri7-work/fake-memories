```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "microsoft/phi-2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

model.eval()
```

```python


prompt = """Alice is the leader.
Bob is the follower.
Charlie is neither.

The leader gives their role to the follower.
The follower becomes the leader.

Who is the leader now?
"""

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        do_sample=False
    )

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
import random
import re
from collections import defaultdict
```

```python
NAME_TRIPLES = [
    ("Alice", "Bob", "Charlie"),
    ("Ravi", "Maya", "Ankit"),
    ("Zeta", "Yuki", "Noor"),
    ("Tom", "Sarah", "Liam"),
    ("Ava", "Ethan", "Iris"),
    ("Neha", "Karan", "Pooja"),
    ("Leo", "Mia", "Noah"),
]

DISTRACTORS = [
    "Cats are mammals.",
    "Dogs bark.",
    "The weather today is sunny.",
    "Water is wet.",
    "Triangles have three sides.",
    "Books contain pages.",
    "Cars have wheels.",
    "Fire is hot.",
]
```

```python
def build_prompt(names, distractors):
    leader, follower, neither = names

    lines = [
        f"{leader} is the leader.",
        f"{follower} is the follower.",
        f"{neither} is neither.",
        ""
    ]

    for d in distractors:
        lines.append(d)

    lines.extend([
        "",
        "The leader gives their role to the follower.",
        "The follower becomes the leader.",
        "",
        "Who is the leader now?"
    ])

    return "\n".join(lines)
```

```python
def run_model(prompt, max_new_tokens=80):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            eos_token_id=tokenizer.eos_token_id
        )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def extract_final_answer(output):
    text = output.lower()

    # Explicit answer
    match = re.search(r"(leader (now )?is|answer is)\s+(alice|bob|charlie|ravi|maya|ankit|zeta|yuki|noor|tom|sarah|liam|ava|ethan|iris|neha|karan|pooja|leo|mia|noah)", text)
    if match:
        return match.group(3)

    # Fallback: last mentioned name
    names = re.findall(
        r"\b(alice|bob|charlie|ravi|maya|ankit|zeta|yuki|noor|tom|sarah|liam|ava|ethan|iris|neha|karan|pooja|leo|mia|noah)\b",
        text
    )
    return names[-1] if names else None
```

```python
def classify_output(output, correct_leader):
    text = output.lower()
    answer = extract_final_answer(output)

    if answer == correct_leader.lower():
        return "correct"

    if answer is None:
        return "no_answer"

    if text.count("who is the leader") > 1:
        return "looping"

    if "roles" in text or "dictionary" in text:
        return "inconsistent_state"

    return "wrong_binding"
```

```python
NUM_INSTANCES = 30
MAX_DISTRACTORS = 4

results = []

for num_distractors in range(0, MAX_DISTRACTORS + 1):
    for _ in range(NUM_INSTANCES):
        names = random.choice(NAME_TRIPLES)
        distractors = random.sample(DISTRACTORS, num_distractors)

        prompt = build_prompt(names, distractors)
        output = run_model(prompt)

        correct_leader = names[1]  # follower becomes leader
        failure_mode = classify_output(output, correct_leader)

        results.append({
            "distractors": num_distractors,
            "mode": failure_mode
        })
```

```python
summary = defaultdict(lambda: defaultdict(int))

for r in results:
    summary[r["distractors"]][r["mode"]] += 1

print("\n=== PHASE 1 SUMMARY (30 INSTANCES EACH) ===\n")
for d in sorted(summary):
    print(f"Distractors = {d}")
    for mode, count in summary[d].items():
        print(f"  {mode:<20} : {count}")
    print()
```

```python
BASE_BINDING = [
    "Alice is the leader.",
    "Bob is the follower.",
    "Charlie is neither."
]

TRANSFORMATION = [
    "The leader gives their role to the follower.",
    "The follower becomes the leader."
]

QUESTION = "Who is the leader now?"
```

```python
DISTRACTORS = [
    "Cats are mammals.",
    "Dogs bark.",
    "The weather today is sunny.",
    "Water is wet.",
    "Triangles have three sides.",
    "Books contain pages.",
    "Fire is hot.",
]
```

```python
def build_prompt(num_distractors):
    lines = []
    lines.extend(BASE_BINDING)
    lines.append("")

    for d in random.sample(DISTRACTORS, num_distractors):
        lines.append(d)

    lines.extend([
        "",
        *TRANSFORMATION,
        "",
        QUESTION
    ])

    return "\n".join(lines)
```

```python
def run_model(prompt, max_new_tokens=80):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            eos_token_id=tokenizer.eos_token_id
        )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
def extract_answer(text):
    text = text.lower()

    match = re.search(r"(leader (now )?is|answer is)\s+(alice|bob|charlie)", text)
    if match:
        return match.group(3)

    names = re.findall(r"\b(alice|bob|charlie)\b", text)
    return names[-1] if names else None
```

```python
def classify_phase1(output):
    text = output.lower()
    answer = extract_answer(output)

    if answer == "bob":
        return "stable"

    if text.count("who is the leader") > 1:
        return "collapse"

    if answer in ["alice", "charlie"]:
        return "unstable"

    if "roles" in text or "dictionary" in text:
        return "unstable"

    return "unstable"
```

```python
NUM_INSTANCES = 30
MAX_DISTRACTORS = 4

summary = defaultdict(lambda: defaultdict(int))

for d in range(MAX_DISTRACTORS + 1):
    for _ in range(NUM_INSTANCES):
        prompt = build_prompt(d)
        output = run_model(prompt)
        mode = classify_phase1(output)
        summary[d][mode] += 1
```

```python
print("\n=== PHASE 1 (CLEAN) SUMMARY ===\n")
for d in sorted(summary):
    print(f"Distractors = {d}")
    for k in ["stable", "unstable", "collapse"]:
        print(f"  {k:<10}: {summary[d].get(k, 0)}")
    print()
```

```python
def tail_window(text, num_chars=300):
    return text.lower()[-num_chars:]
```

```python
def detect_binding(text):
    bindings = {
        "bob": [
            "bob is the leader",
            "leader is bob",
            "bob becomes the leader",
            "bob is now the leader"
        ],
        "alice": [
            "alice is the leader",
            "leader is alice",
            "alice becomes the leader",
            "alice is now the leader"
        ]
    }

    found = set()
    for name, patterns in bindings.items():
        for p in patterns:
            if p in text:
                found.add(name)
    return found
```

```python
def classify_phase1B(output):
    text = output.lower()

    # Collapse = looping / no termination
    if text.count("who is the leader") > 1:
        return "collapse"

    tail = tail_window(text)
    bindings = detect_binding(tail)

    if bindings == {"bob"}:
        return "stable"

    if "bob" in bindings and "alice" in bindings:
        return "unstable"  # contradiction

    if bindings:
        return "unstable"

    return "unstable"
```

```python
NUM_INSTANCES = 30
MAX_DISTRACTORS = 4
OUTPUTS=[]

summary = defaultdict(lambda: defaultdict(int))

for d in range(MAX_DISTRACTORS + 1):
    for _ in range(NUM_INSTANCES):
        prompt = build_prompt(d)
        output = run_model(prompt)
        OUTPUTS.append(output)

        mode = classify_phase1(output)
        summary[d][mode] += 1
```
